```python
import pandas as pd
homelessness = pd.DataFrame([['East South Central', 'Alabama', 2570, 864, 4887681],
                             ['Pacific', 'Alaska', 1434, 582, 735139],
                             ['Mountain', 'Arizona', 7259, 2606, 7025024],
                             ['West South Central', 'Arkansas', 2280, 432, 3009733],
                             ['Pacific', 'California', 109008, 20964, 39461588]],
                            columns=['region', 'state', 'individuals', 'family_members', 'state_pop'])

print(homelessness.head())
print(homelessness.info())
print(homelessness.shape)
print(homelessness.describe())

print(homelessness.columns)
print(homelessness.index)
print(homelessness.values)

print(homelessness.sort_values('region'))
print(homelessness.sort_values(['individuals', 'region'], ascending=[False, True]))

print(homelessness[['state', 'individuals']])
ind_lessthan = homelessness['individuals'] < 10000
famm_morethan = homelessness['family_members'] > 500
print(homelessness[ind_lessthan & famm_morethan])
print(homelessness[(homelessness['individuals']<10000) & (homelessness['family_members']>500)])

is_in_state = homelessness['state'].isin(['Alabama', 'Arizona'])
print(homelessness[is_in_state])

def pct30(column):
    return column.quantile(0.3)
def pct40(column):
    return column.quantile(0.4)
print(homelessness[['individuals', 'state_pop']].agg([pct30, pct40]))

sales = pd.DataFrame([[1, 'A', 1, pd.to_datetime('2010-02-05'), 24924.50, False, 5.728, 0.679, 8.106],
                      [1, 'A', 2, pd.to_datetime('2010-03-05'), 21827.90, False, 8.056, 0.693, 8.106],
                      [2, 'A', 1, pd.to_datetime('2010-04-02'), 57258.43, False, 16.817, 0.718, 7.808],
                      [2, 'A', 2, pd.to_datetime('2010-05-07'), 17413.94, False, 22.528, 0.749, 7.808],
                      [2, 'A', 2, pd.to_datetime('2010-06-04'), 17558.09, False, 27.050, 0.715, 7.808]],
                      columns=['store','type', 'department', 'date', 'weekly_sales', 'is_holiday', 'temperature_c', 'fuel_price_usd_per_l', 'unemployment'])

sales_A = sales[sales['type']=='A']['weekly_sales'].sum()
sales_all = sales.groupby(['type', 'is_holiday'])['weekly_sales'].sum()
print(sales_A)
print(sales_all)

import numpy as np
print(sales.groupby('type')[['unemployment', 'fuel_price_usd_per_l']].agg([min, max, np.mean, np.median]))

print(sales.pivot_table(values='weekly_sales', index='type', aggfunc=[min, max, np.mean, np.median], margins=True))

sales_ind = sales.set_index(['store', 'date'])
print(sales_ind)
print(sales_ind.sort_index(level=['date', 'store'], ascending=[False, False]))
rows_to_keep = [(1, '2010-02-05'), (1, '2010-03-05'), (2, '2010-04-02')]
print(sales_ind.loc[rows_to_keep])
print(sales_ind.iloc[1:5, 2:3])

#Boolean conditions slicing can ONLY use full dates:
sales_bool = sales[(sales['date'] >= '2010-01-01') & (sales['date'] <= '2010-05-30')]
print(sales_bool['date'])

#Indexes CAN use trimmed dates:
sales_ind = sales.set_index('date').sort_index()
print(sales_ind.loc['2010-02':'2010-05'])
#WARNING! Not specyfying explicitly dtype introduces some peculiarities. For example, impossibility of:
print(sales.iloc[3,3].month)
print(sales['date'].dt.year)
```
